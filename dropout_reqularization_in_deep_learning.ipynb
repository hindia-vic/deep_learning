```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#getting data from uci sonar dataset
data=pd.read_csv('/content/sample_data/sonar.all-data',header=None)
data.head()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
X.head()
```

```python
#one hot encoding  for my y
y=pd.get_dummies(y,drop_first=True,dtype=int)
y.head()
```

```python
y.value_counts()
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

```

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
#building the model with dropout layer that will drop 0.5
modeld=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid')
])
modeld.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
modeld.fit(X_train,y_train,epochs=100,batch_size=8)
```

```python
modeld.evaluate(X_test,y_test)
```

```python
y_pred=modeld.predict(X_test).reshape(-1)
print(y_pred[:10])
y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
print(classification_report(y_test,y_pred))
```
